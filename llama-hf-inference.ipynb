```python
import torch
from peft import PeftModel, PeftConfig
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig

peft_model_id = "MonishMeher/catalyst-kiran"
config = PeftConfig.from_pretrained(peft_model_id)

model = AutoModelForCausalLM.from_pretrained(
    config.base_model_name_or_path,
    #quantization_config=bnb_config,
    torch_dtype=torch.bfloat16,
    device_map={"":0},
    trust_remote_code=True
)
tokenizer = AutoTokenizer.from_pretrained(config.base_model_name_or_path)
tokenizer.pad_token = tokenizer.eos_token

# Load the Lora model
model = PeftModel.from_pretrained(model, peft_model_id)
```

```python
text = "who is Steve Jobs?"
```

### Low Repetition Penalty (e.g., 1.0): The model is not penalized much for repeating tokens. This might lead to more repetitive output, as the model is free to select tokens it has already generated.

### Higher Repetition Penalty (e.g., >1.0): The model is penalized more for repeating tokens. This encourages the model to produce more diverse and varied output by discouraging the repetition of the same words or phrases.


```python
device = "cuda:0"

inputs = tokenizer(text, return_tensors="pt").to(device)
outputs = model.generate(**inputs, repetition_penalty=1.2, max_new_tokens=64)
```

```python
print(tokenizer.decode(outputs[0], skip_special_tokens=True))
```

## Use Model After Training

Once you have trained a model using either the `SFTTrainer`, `PPOTrainer`, or `DPOTrainer`, you will have a fine-tuned model that can be used for text generation.
In this section, we’ll walk through the process of loading the fine-tuned model and generating text. If you need to run an inference server with the trained model,
you can explore libraries such as [text-generation-inference](https://github.com/openai/text-generation-inference).


### Load and Generate

If you have fine-tuned a model fully, meaning without the use of PEFT, you can simply load it like any other language model in transformers.
For example, the value head that was trained during the PPO training is no longer needed, and if you load the model with the original transformer class, it will be ignored:

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

model_name_or_path = "MonishMeher/catalyst-kiran" #path/to/your/model/or/name/on/hub
device = "cuda" # or "cuda" if you have a GPU

model = AutoModelForCausalLM.from_pretrained(model_name_or_path).to(device)
tokenizer = AutoTokenizer.from_pretrained(model_name_or_path)

inputs = tokenizer.encode("This movie was really", return_tensors="pt").to(device)
outputs = model.generate(inputs)
print(tokenizer.decode(outputs[0]))
```

```python
model = model.merge_and_unload()
```
